▪︎`pandas`, `scipy.stats`, `matplotlib.pyplot`モジュールをそれぞれ，`pd`, `stats`, `plt`という名前でインポートする：

````python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
````

▪︎`pandas`モジュールの`read_csv`関数を用いて，センサデータCSVファイル`'memdata_TSND121-06120699_2019-04-12_140934.329.csv'`を読み込み，変数`df`に格納する：

````python
df = pd.read_csv('memdata_TSND121-06120699_2019-04-12_140934.329.csv',
                 names=['item', 'time', 'acc_x', 'acc_y', 'acc_z', 'omega_x', 'omega_y', 'omega_z'])
````

> センサデータCSVファイルには項目名がないため，オプション引数
```
names=['item', 'time', 'acc_x', 'acc_y', 'acc_z', 'omega_x', 'omega_y', 'omega_z']
```
を指定して，データフレーム`df`に項目名を設定している。

▪︎`df`の最上部と最下部を表示してみる：

````python
df.head()
````

````python
df.tail()
````

▪︎データの最初と最後の部分は怪しい（撹乱が入る場合がある）ので，`df`の1000~15000行目までを残し，後はカットすることにした：

````python
df = df[1000:15001]
````

データフレームオブジェクトのメソッド`describe()`で，データの統計的要約を見てみる：

````python
df.describe()
````

> ここで，
>
> | 要約名称 | 意味 |
|---|---|
| count | 個数 |
| mean | 平均値 |
| std | 標準偏差 |
| min | 最小値 |
| 25% | 第１四分位数 |
| 50% | 第２四分位数(中央値) |
| 75% | 第３四分位数 |
| max | 最大値 |

▪︎データフレームオブジェクトの`hist(...)`メソッドを用いて，`df`のすべての項目についてヒストグラムを描く:

````python
df.hist(figsize=(20, 20),bins=20)
plt.show()
````

````python
for i in range(2,8):
    plt.figure(figsize=(5,5))
    stats.probplot(df.iloc[:, i], dist="norm", plot=plt)
    plt.show()
````
